```python
df2 = pd.read_csv('edhec-hedgefundindices.csv')
df2
```

```python
df2['date'] = pd.to_datetime(df2['date'])

# Filter the df2 for the years 2000 to 2018
data_filtered = df2[(df2['date'].dt.year >= 2000) & (df2['date'].dt.year <= 2018)]

# Remove the 'date' column for easier calculations
data_filtered = data_filtered.set_index('date')

# Calculate the annualized return for each portfolio
annualized_returns = {}

for column in data_filtered.columns:
    # Convert monthly returns from percentages if necessary
    monthly_returns = data_filtered[column] / 100  # Remove this if already in decimal form
    cumulative_return = (1 + monthly_returns).prod() - 1
    number_of_months = len(monthly_returns)
    annualized_return = (1 + cumulative_return) ** (12 / number_of_months) - 1
    annualized_returns[column] = annualized_return

# Find the portfolio with the lowest annualized return
lowest_portfolio = min(annualized_returns, key=annualized_returns.get)
lowest_return = annualized_returns[lowest_portfolio]

print(f"The portfolio with the lowest annualized return from 2000-2018 is: {lowest_portfolio}")
print(f"Annualized Return: {lowest_return * 100:.2f}%")
```

```python
from scipy.stats import shapiro
lo_10_returns = df['Lo 10'].dropna()
hi_10_returns = df['Hi 10'].dropna()

# Perform the Shapiro-Wilk test for normality
lo_10_stat, lo_10_p = shapiro(lo_10_returns)
hi_10_stat, hi_10_p = shapiro(hi_10_returns)

# Define significance level
alpha = 0.05

# Print the results
if lo_10_p > alpha:
    print("The 'Lo 10' portfolio returns may follow a normal distribution (p-value:", lo_10_p, ")")
else:
    print("The 'Lo 10' portfolio returns do not follow a normal distribution (p-value:", lo_10_p, ")")

if hi_10_p > alpha:
    print("The 'Hi 10' portfolio returns may follow a normal distribution (p-value:", hi_10_p, ")")
else:
    print("The 'Hi 10' portfolio returns do not follow a normal distribution (p-value:", hi_10_p, ")")
```

```python
df2['date'] = pd.to_datetime(df2['date'])

# Filter df2 for the years 2009 to 2018
data_filtered = df2[(df2['date'].dt.year >= 2009) & (df2['date'].dt.year <= 2018)]

# Set the 'date' column as the index
data_filtered = data_filtered.set_index('date')

# Function to calculate semi-deviation (downside risk) for each column
def semi_deviation(returns):
    mean_return = returns.mean()
    downside_returns = returns[returns < mean_return]
    return np.std(downside_returns, ddof=1)  # ddof=1 for sample standard deviation

# Calculate semi-deviation for each hedge fund index
semi_deviation_values = {column: semi_deviation(data_filtered[column].dropna()) for column in data_filtered.columns}

# Find the index with the highest semi-deviation
highest_semi_deviation_index = max(semi_deviation_values, key=semi_deviation_values.get)
highest_semi_deviation_value = semi_deviation_values[highest_semi_deviation_index]

# Print the result
print(f"The Hedge Fund Index with the highest semi-deviation from 2009 to 2018 is: {highest_semi_deviation_index}")
print(f"Semi-deviation value: {highest_semi_deviation_value}")
```
